```python
## Creating a function : price_label

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df_subset['price_range'] = df_subset.apply(price_label, axis=1)
```

```python
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
## Value 0 means saturday, Wednesday(4) has the lowest frecuency (slowest day)
```

```python
## Creating a new column (busiest day) The loop will run through every row in the "orders_day_of_week"column
## assigning corresponding string value.

result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
## Checking the new result columnm
result
```

```python
## Creating a new column withing the dataframe and setting it equal to result
df_ords_prods_merge['busiest_day'] = result
```

```python
## Printing the frecuency of this new columnn
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
## Step 2
#Creating new column busiests_days

result2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value in [0, 1]:
    result2.append("Busiests days")
  elif value in [4, 3]:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
 
 
```
