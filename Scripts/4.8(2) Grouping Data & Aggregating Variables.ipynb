```python
##Step 6. Creating a new flag
#### Splitting the data into groups based on user_id.
## Applying the transform() function on the prices columnn to generate the mean orders for each user.
##Create a new column,"mean_prices" into which you'll place the results of your aggregation.

# Calculate the mean price per user
df_ords_prods['mean_prices'] = df_ords_prods.groupby('user_id')['prices'].transform('mean')
```

```python
## for checking all 100 rows
pd.options.display.max_rows = None
```

```python
df_ords_prods.head(100)
```

```python
## Deriving Columns with loc()
df_ords_prods.loc[df_ords_prods['mean_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods.loc[df_ords_prods['mean_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
## printing
df_ords_prods['spending_flag'].value_counts(dropna = False)
```

```python
## Step7.
# Calculate median days_since_prior_order per user
df_ords_prods['median_days_since_prior_order'] = df_ords_prods.groupby('user_id')['days_since_prior_order'].transform('median')
```

```python
# Make sure df_ords_prods is a copy
df_ords_prods = df_ords_prods.copy()
```

```python
## Step7.
# Calculate median days_since_prior_order per user
df_ords_prods['median_days_since_prior_order'] = df_ords_prods.groupby('user_id')['days_since_prior_order'].transform('median')
```

```python
# Deriving columns with loc()
df_ords_prods.loc[df_ords_prods['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods.loc[
    (df_ords_prods['median_days_since_prior_order'] > 10) & (df_ords_prods['median_days_since_prior_order'] <= 20), 
    'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods.loc[df_ords_prods['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
